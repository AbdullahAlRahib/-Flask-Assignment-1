# **Q3**

**What is App routing in Flask? Why do we use app routes?**


**Answer:**

App routing in Flask refers to the process of mapping URL patterns to specific functions or view handlers within a Flask application. It defines how the application responds to different HTTP requests for different URLs.

In Flask, app routes are defined using the @app.route() decorator, which associates a URL pattern with a Python function. This decorator allows developers to specify the URL endpoint and the HTTP methods that the function should handle.

We use app routes in Flask for several reasons:

 * URL Mapping.

 * Request Handling.

 * Dynamic URLs.

 * RESTful APIs.

 * Separation of Concerns.

 * URL Generation.

**Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:**

Company Name: ABC Corporation

Location: India

# **Q5**

**What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function..**

**Answer:**


url_for() function is used for URL building. It generates a URL for a specific endpoint based on its name and any additional arguments provided.



```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, Flask!'

@app.route('/user/<username>')
def user_profile(username):
    return f'Profile page of {username}'

@app.route('/post/<int:post_id>')
def show_post(post_id):
    return f'Post #{post_id}'

@app.route('/about')
def about():
    return 'About page'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('index'))  # Output: /
        print(url_for('user_profile', username='john'))  # Output: /user/john
        print(url_for('show_post', post_id=123))  # Output: /post/123
        print(url_for('about'))  # Output: /about
```

In the above code, we have defined several routes using the @app.route() decorator. The url_for() function is then used to generate URLs for these routes. The function takes the name of the endpoint as the first argument and any additional arguments required by the endpoint. In this example, we demonstrate generating URLs for the index, user_profile, show_post, and about endpoints.

When running the code, we use the url_for() function within the test_request_context() to simulate a request context and print the generated URLs. The output shows the corresponding URLs for each endpoint based on the provided arguments.
